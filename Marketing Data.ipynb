# VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting the aesthetic style of the plots

sns.set_style('whitegrid')
```

```python
# Plotting distributions of the advertising spending and sales

tqdm.pandas()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(Data["youtube"], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of YouTube Advertising Spending")

sns.histplot(Data["facebook"], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Distribution of Facebook Advertising Spending")

sns.histplot(Data["newspaper"], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Distribution of Newspaper Advertising Spending")

sns.histplot(Data["sales"], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Distribution of Sales")

plt.tight_layout()
plt.show()
```

```python
# The first plot shows the distribution of YouTube advertising spending.
# The second plot shows the distribution of Facebook advertising spending.
# The third plot shows the distribution of Newspaper advertising spending.
# The fourth plot shows the distribution of sales.

## These visualizations can help in understanding how the spending on different advertising platforms is distributed and how it
#  relates to sales. 
```

# Creating scatter plots to visualize the relationships between the advertising spending on different platforms and the sales and including a regression line to indicate trends

```python
# This will help us understand if there is a correlation between advertising spending and sales.
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.regplot(x="youtube", y="sales", data=Data, ax=axes[0])
axes[0].set_title("YouTube Spending vs Sales")

sns.regplot(x="facebook", y="sales", data=Data, ax=axes[1])
axes[1].set_title("Facebook Spending vs Sales")

sns.regplot(x="newspaper", y="sales", data=Data, ax=axes[2])
axes[2].set_title("Newspaper Spending vs Sales")

plt.tight_layout()
plt.show()
```

```python
# The regression lines suggest trends that could be explored further to understand the effectiveness of each advertising platform
```

# Correlation Analysis

```python
#  This will provide a clearer picture of how each type of spending is associated with sales outcomes
```

```python
# Calculate the correlation matrix to quantify the relationship between advertising spending and sales.

correlation_matrix = Data.corr()
```

```python
correlation_matrix
```

```python
# Visualize the correlation matrix using a heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Advertising Spending and Sales")
plt.show()
```

```python
# YouTube advertising has a strong positive correlation with sales (r = 0.78).
# Facebook advertising also shows a positive correlation with sales (r = 0.60).
# Newspaper advertising has the weakest positive correlation with sales (r = 0.25).

# These results suggest that YouTube and Facebook advertising spending is more strongly associated with increased sales 
# compared to Newspaper advertising.

# This analysis can guide marketing budget allocation decisions indicating that YouTube and Facebook may offer better returns
# on investment in terms of sales than the newspaper.
```

# Exploring the potential return on investment (ROI) for each advertising platform.

```python
# comparing the cost of advertising on each platform with the sales generated from that spending

# This gives an insight into which platform yields the highest sales return for each dollar spent.
```

```python
# Calculate the total and average spending for each platform

total_spending = Data[["youtube", "facebook", "newspaper"]].sum()
```

```python
average_spending = Data[["youtube", "facebook", "newspaper"]].mean()
```

```python
# Calculate the total and average sales
total_sales = Data["sales"].sum()
```

```python
average_sales = Data["sales"].mean()
```

```python
# Calculate the ROI for each platform

roi_youtube = total_sales / total_spending["youtube"]
```

```python
roi_facebook = total_sales / total_spending["facebook"]
```

```python
roi_newspaper = total_sales / total_spending["newspaper"]
```
