# These datasets collectively offer a comprehensive view of drug use, obesity admissions and smoking across different demographics and regions for the year 2019/20. To further analyze these files and visualize the data, plotting charts of distributions and trends based on the provided counts, rates, and admissions is needed

```python
# Plotting distributions for drug use by age group

plt.figure(figsize=(14, 7))
age.groupby("AgeGroup")["Count"].sum().plot(kind="bar", color="green")
plt.title("Drug Use by Age Group (2019/20)")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Drug Use by Age Group (2019/20): The bar chart above shows the count of drug use cases across different age groups.
```

```python
# Plotting distributions for drug use by ICD10 diagnosis

plt.figure(figsize=(14, 7))
diag.groupby("ICD10_Description")["Count"].sum().plot(kind="bar", color="skyblue")
plt.title("Drug Use by ICD10 Diagnosis (2019/20)")
plt.xlabel("ICD10 Diagnosis")
plt.ylabel("Count")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# Drug Use by ICD10 Diagnosis (2019/20): This bar chart(above) displays the count of drug use cases categorized by ICD10 diagnosis
# descriptions.
```

```python
# Plotting distributions for obesity admissions by age group

plt.figure(figsize=(14, 7))
obesity_age.groupby("AgeGroup")["Admissions"].sum().plot(kind="bar", color="purple")
plt.title("Obesity Admissions by Age Group (2019/20)")
plt.xlabel("Age Group")
plt.ylabel("Admissions")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Obesity Admissions by Age Group (2019/20): This bar chart(above) represents the number of obesity admissions across various 
# age groups.
```

```python
# Smoking admissions distribution

smoking_admissions["Metric_Primary"].plot(kind="hist", bins=20, alpha=0.7)
plt.title("Smoking Admissions Distribution")
plt.xlabel("Metric Primary")
plt.ylabel("Frequency")
plt.show()
```

```python
# Inspect the 'Metric_Primary' column of the smoking admissions dataframe

df = smoking_admissions["Metric_Primary"]
print(df.head())
print(df.describe())
print(df.dtypes)
```

```python
# The "Metric_Primary" column in the smoking admissions dataframe is not numeric; it contains categorical data with the value
# "Number of admissions" repeated throughout the column with object as its data type. Histograms require numeric data(int)
```

```python
# Checking the smoking admissions dataframe to find the column with numbers

print(smoking_admissions.head())
print(smoking_admissions.describe(include="all"))
```

```python
# The correct column with numbers in the smoking admissions dataframe that represents the count or number of admissions is the 
# "Value" column. This column contains the data for the number of admissions.
```

```python
# Plotting the histogram for the 'Value" column which contains the number of admissions

plt.figure(figsize=(10, 6))
smoking_admissions["Value"].plot(kind="hist", bins=30, alpha=0.7)
plt.title("Smoking Admissions Distribution")
plt.xlabel("Number of Admissions")
plt.ylabel("Frequency")
plt.show()
```

```python
# The chart above provides a frequency distribution of the number of admissions across different counts.
```
